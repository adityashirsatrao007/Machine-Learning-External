```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Generate some sample data
X = np.random.rand(50, 1) * 10  # 50 random numbers between 0 and 10
y = 2 * X + 1 + np.random.randn(50, 1) * 2 # y = 2x + 1 + some noise

# 2. Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# 3. Make predictions for the regression line
y_pred = model.predict(X)

# 4. Plot the data and the regression line
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.title('Simple Linear Regression')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```
