```python
!pip install ucimlrepo
from ucimlrepo import fetch_ucirepo

# fetch dataset
student_performance = fetch_ucirepo(id=320)

# data (as pandas dataframes)
X = student_performance.data.features
y = student_performance.data.targets

# metadata
print(student_performance.metadata)

# variable information
print(student_performance.variables)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

y.loc[:, 'G3_classified'] = np.where(y['G3'] >= 10, 1, 0)
X_train, X_test, y_train, y_test = train_test_split(X, y['G3_classified'], test_size=0.2, random_state=42, stratify=y['G3_classified'])

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

sns.countplot(x=y_train, ax=axes[0], palette='viridis')
axes[0].set_title('Training Set Distribution')
axes[0].set_xlabel('G3_classified')
axes[0].set_ylabel('Count')

sns.countplot(x=y_test, ax=axes[1], palette='plasma')
axes[1].set_title('Testing Set Distribution')
axes[1].set_xlabel('G3_classified')
axes[1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
