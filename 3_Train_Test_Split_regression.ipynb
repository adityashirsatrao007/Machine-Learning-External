```python
!pip install ucimlrepo
from ucimlrepo import fetch_ucirepo

# fetch dataset
student_performance = fetch_ucirepo(id=320)

# data (as pandas dataframes)
X = student_performance.data.features
y = student_performance.data.targets

# metadata
print(student_performance.metadata)

# variable information
print(student_performance.variables)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'X' contains features and 'y' contains targets from previous steps
# Select 'G3' as the target variable for regression
y_g3 = y['G3']

# Perform the train-test split
# test_size=0.2 means 20% of the data will be used for testing
# random_state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y_g3, test_size=0.2, random_state=42)

print(f"Training set size: {len(X_train)} samples")
print(f"Testing set size: {len(X_test)} samples")

# Create a single plot to visualize the distribution of G3 in train and test sets
plt.figure(figsize=(10, 6))
sns.histplot(y_train, color='blue', label='Train G3', kde=True, stat='density', alpha=0.5)
sns.histplot(y_test, color='red', label='Test G3', kde=True, stat='density', alpha=0.5)
plt.title('Distribution of Final Grades (G3) in Training and Testing Sets')
plt.xlabel('Final Grade (G3)')
plt.ylabel('Density')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
